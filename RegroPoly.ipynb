```python
def zillowStateSeries(file,state):
    """returns an array from zillow csv data
     """
    return pd.read_csv(file, encoding='ISO-8859-1').drop(columns=['RegionID','SizeRank'])[state]
```

```python
def zillowStateDF(file,state):
    """returns a df from zillow csv data
     """
    return pd.read_csv(file, encoding='ISO-8859-1').drop(columns=['RegionID','SizeRank'])
```

```python
def zillowStatePreGameDF(file,state,date):
    """returns a df from zillow csv data
     """
    df=pd.read_csv(file, encoding='ISO-8859-1').drop(columns=['RegionID','SizeRank'])
    return df
```
